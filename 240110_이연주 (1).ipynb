```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
import os
```

```python
folder_path = 'image'
img_paths = os.listdir(folder_path)
```

```python
# 구글 코랩에서는 cv2.imshow 가 안됨
from google.colab.patches import cv2_imshow

img_stack = []
for i, img_path in enumerate(img_paths):
  img = cv2.imread(os.path.join(folder_path, img_path))
  img = cv2.resize(img, (300, 400))
  gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
  gray_img = cv2.cvtColor(gray_img, cv2.COLOR_GRAY2BGR)
  if i == 0:
    img_stack = np.concatenate((img, gray_img), axis=1)
  else:
    img_stack = np.concatenate((img_stack, (np.concatenate((img, gray_img), axis=1))), axis=0)
cv2_imshow(img_stack)
```
